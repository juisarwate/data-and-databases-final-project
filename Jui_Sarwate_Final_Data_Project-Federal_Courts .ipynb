```python
import requests
from bs4 import BeautifulSoup
import json
import numpy as np
import pandas as pd
from pandas import json_normalize
```

```python
my_url = "http://www.uscourts.gov/judges-judgeships/judicial-vacancies/confirmation-listing"
raw_html = requests.get(my_url).content

good_soup = BeautifulSoup(raw_html, "html.parser")
print(good_soup.prettify())
```

```python
rows = good_soup.find_all('tr')
all_rows = []
for row in rows[1:]:
    myrow = []
    info = row.find_all('td')
    for data in info:
        myrow.append(data.string)
    all_rows.append(myrow)
all_rows
```

```python
col_names = ['nominee', 'nomination_date', 'confirmation_date', 'court','incumbent','vacancy_reason','vacancy_date']
df = pd.DataFrame.from_records(all_rows, columns=col_names)
df.head()
```

```python
df.to_csv(r'backup_confirmations.csv', index = False)
```

```python
# Get out the ABBR of the court
df[['court_id', 'properties_ABBR']] = df['court'].str.split('-', 1, expand=True)
df
```

```python
nominee = df.properties_ABBR
pd.DataFrame(nominee)
```

```python
# Groupby the properties.ABBR and turn into dataframe
courts = df.groupby('properties.ABBR')['nominee'].nunique()
df_courts = pd.DataFrame(courts)
df_courts
```
